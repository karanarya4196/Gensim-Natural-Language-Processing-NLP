```python
import pandas as pd
import nltk
import gensim
from gensim import corpora, models, similarities
```

```python
my_file = open("2001-Bush.txt")
my_file_str = my_file.read()
```

```python
words = nltk.word_tokenize(my_file_str)
```

```python
words
```

```python
model = gensim.models.Word2Vec([words], min_count = 1, size = 100)
```

```python
print("Cosine similarity between 'citizens' " + "and 'democracy' - CBOW : ", model.similarity('citizens', 'democracy')) 
```

```python
model2 = gensim.models.Word2Vec([words], min_count = 1, size = 100, window = 5, sg = 1)
```

```python
print("Cosine similarity between 'citizens' " + "and 'democracy' - Skip-Gram : ", model2.similarity('citizens', 'democracy')) 
```
